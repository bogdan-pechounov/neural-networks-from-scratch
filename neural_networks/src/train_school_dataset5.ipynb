```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/student/student-mat.csv', sep=';')
df
```

```python
pd.get_dummies(df)
```

```python
df_exclude = df[df.columns[~df.columns.isin(['G1','G2', 'G3'])]] 
df_categorical = pd.get_dummies(df_exclude)
X = df_categorical.to_numpy()
Y = df.filter(items=['G1', 'G2', 'G3']).to_numpy()
```

```python
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from neural_network import NeuralNetwork
from loss.mse import MSE
from layer import Layer
from activation.tanh import Tanh
from activation.relu import ReLU


nn = NeuralNetwork(MSE)
nn.add(Layer(56, 10))
nn.add(Tanh())
nn.add(Layer(10, 5))
nn.add(Tanh())
nn.add(Layer(5, 3))


print(nn)
losses = nn.train(X_train, Y_train, epochs=200000, learning_rate=0.001)
print(nn)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from loss.mae import MAE

plt.ylabel('Train RMSE')
plt.plot(np.sqrt(losses))
plt.show()

prediction = nn.forward(X_test)
mse = MSE.loss(prediction, Y_test)
rmse = np.sqrt(mse)
mae = MAE().loss(prediction, Y_test)
print('train:', MAE().loss(nn.forward(X_train), Y_train), np.sqrt(MSE.loss(nn.forward(X_train), Y_train)))
print('test:', mae, rmse)
```
